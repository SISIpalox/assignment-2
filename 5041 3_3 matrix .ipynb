```python
from itertools import permutations
import numpy as np
from numpy.linalg import det

# Generate all permutations of the array with five 0s and four 1s
all_permutations_3x3 = set(permutations([0, 0, 0, 0, 0, 1, 1, 1, 1]))

# Convert each permutation into a 3x3 matrix and calculate its determinant
matrices_with_determinants_3x3 = []

for perm in all_permutations_3x3:
    matrix = np.array(perm).reshape(3, 3)
    determinant = det(matrix)

    matrices_with_determinants_3x3.append((matrix, determinant))


print(f"Number of unique matrices: {len(matrices_with_determinants_3x3)}")


for matrix, determinant in matrices_with_determinants_3x3:
    print("Matrix:\n", matrix)
    print("Determinant:", determinant)
    print() 
```

```python
from itertools import permutations
import numpy as np

def count_matrices_with_zero_determinant_five_zeros():
    # Generate all permutations of five 0s and four 1s for the 3x3 matrix
    all_permutations = set(permutations([0, 0, 0, 0, 0, 1, 1, 1, 1]))
    zero_det_count = 0

    for perm in all_permutations:
        # Construct the matrix from the permutation
        matrix = np.array(perm).reshape(3, 3)
        # Calculate the determinant
        if np.linalg.det(matrix) == 0:
            zero_det_count += 1

    return zero_det_count

# Calculate the number of matrices with a determinant of 0 using five 0s and four 1s
matrices_with_zero_det_five_zeros = count_matrices_with_zero_determinant_five_zeros()
print(matrices_with_zero_det_five_zeros)
```

```python
from itertools import permutations
import numpy as np

def find_most_common_zero_position():
    # Generate all permutations of five 0s and four 1s for the 3x3 matrix
    all_permutations = set(permutations([0, 0, 0, 0, 0, 1, 1, 1, 1]))
    zero_positions_count = np.zeros((3, 3))

    for perm in all_permutations:
        # Construct the matrix from the permutation
        matrix = np.array(perm).reshape(3, 3)
        # Calculate the determinant
        if np.linalg.det(matrix) == 0:
            # Increment the count for positions where zero appears
            zero_positions_count += (matrix == 0)

    # Find the position where zero appears most frequently in matrices with determinant 0
    most_common_zero_position = np.unravel_index(np.argmax(zero_positions_count), zero_positions_count.shape)
    most_common_zero_count = np.max(zero_positions_count)

    return most_common_zero_position, most_common_zero_count

# Find the position of zero that appears most in all 0 determinant cases
most_common_zero_position, most_common_zero_count = find_most_common_zero_position()

print("Most common zero position:", most_common_zero_position)
print("Count:", most_common_zero_count)
```
